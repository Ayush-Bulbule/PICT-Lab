```python
import nltk
nltk.download('all')
```

```python
document="NLP stands for Natural Language Processing. It is the elective subject in eighth semester."
```

```python
from nltk import sent_tokenize
from nltk import word_tokenize

tokens=word_tokenize(document)
tokens
```

```python
sent_tokens=sent_tokenize(document)
sent_tokens
```

```python
from nltk import WhitespaceTokenizer

wh_tokens=WhitespaceTokenizer().tokenize(document)
wh_tokens
```

```python
from nltk.tokenize import MWETokenizer

mwet_tokens=MWETokenizer().tokenize(document.split())
mwet_tokens
```

```python
from nltk.tokenize import TweetTokenizer

tweet_tokens=TweetTokenizer().tokenize(document)
tweet_tokens
```

```python
from nltk.tokenize import TreebankWordTokenizer

tb_tokens=TreebankWordTokenizer().tokenize(document)
tb_tokens
```

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

for w in tokens:
    print(stemmer.stem(w))
```

```python
from nltk.stem import SnowballStemmer

sn_stemmer=SnowballStemmer(language="english")

for w in tokens:
  print(sn_stemmer.stem(w))
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer=WordNetLemmatizer()

for w in tokens:
  print(lemmatizer.lemmatize(w))
```
