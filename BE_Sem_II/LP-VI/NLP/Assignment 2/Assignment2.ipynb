```python
import pandas as pd
```

```python
import re
```

```python
from nltk.tokenize import word_tokenize
```

```python
from nltk.corpus import stopwords
```

```python
df=pd.read_csv("data.csv")
df
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
def preprocess_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'\W', ' ', text)  # Remove special characters
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces
    tokens = word_tokenize(text)  # Tokenization
    tokens = [word for word in tokens if word not in stopwords.words('english')]  # Remove stopwords
    return " ".join(tokens)

# Apply preprocessing
df['cleaned_text'] = df["Market Category"].astype(str).apply(preprocess_text)

# Display cleaned text
df[['Market Category', 'cleaned_text']].head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
vectorizer = CountVectorizer()
X_bow = vectorizer.fit_transform(df['cleaned_text'])

# Convert to DataFrame
bow_df = pd.DataFrame(X_bow.toarray(), columns=vectorizer.get_feature_names_out())

# Display BoW representation
bow_df.head()
```

```python
from sklearn.preprocessing import normalize

# Convert sparse matrix to dense before normalization
X_bow_dense = X_bow.toarray()

# Apply L1 normalization
X_bow_normalized = normalize(X_bow_dense, norm='l1', axis=1)  # L1 normalization (row-wise)

# Convert to DataFrame
bow_normalized_df = pd.DataFrame(X_bow_normalized, columns=vectorizer.get_feature_names_out())

# Display normalized BoW representation
bow_normalized_df.head()
```

```python
tfidf_vectorizer = TfidfVectorizer()
X_tfidf = tfidf_vectorizer.fit_transform(df['cleaned_text'])

# Convert to DataFrame
tfidf_df = pd.DataFrame(X_tfidf.toarray(), columns=tfidf_vectorizer.get_feature_names_out())

# Display TF-IDF representation
tfidf_df.head()
```

```python
from gensim.models import Word2Vec
```

```python
tokenized_sentences = [word_tokenize(text) for text in df['cleaned_text']]

# Train Word2Vec model
word2vec_model = Word2Vec(sentences=tokenized_sentences, vector_size=100, window=5, min_count=1, workers=4)

# Get word vectors
word_vectors = word2vec_model.wv

# Example: Display vector for a word
word = "diesel"
if word in word_vectors:
    print(f"Vector for '{word}':\n", word_vectors[word])
else:
    print(f"'{word}' not found in vocabulary.")
```
