```python
import pandas as pd
```

```python
df=pd.read_pickle("News_dataset.pickle")
df
```

```python
from nltk.tokenize import word_tokenize
```

```python
df["tokens"]=df["Content"].astype(str).apply(word_tokenize)
df
```

```python
from nltk.stem import WordNetLemmatizer
wnl = WordNetLemmatizer()

df["lemma"] = df["tokens"].apply(lambda tokens: [wnl.lemmatize(word) for word in tokens])
df["lemma"]
```

```python
from nltk.corpus import stopwords
st=list(stopwords.words('english'))

def remove_stopwords(tokens):
    return [word for word in tokens if word.lower() not in st]

df["filtered_lemma"] = df["lemma"].apply(remove_stopwords)
df["filtered_lemma"]
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['category_code']=le.fit_transform(df['Category'])
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pickle

# Convert lemmatized tokens back to text
df["processed_text"] = df["filtered_lemma"].apply(lambda tokens: " ".join(tokens))

# Initialize TF-IDF Vectorizer
vectorizer = TfidfVectorizer()

# Fit and transform the text data
tfidf_matrix = vectorizer.fit_transform(df["processed_text"])

# Save the TF-IDF matrix and vectorizer
with open("tfidf_matrix.pkl", "wb") as f:
    pickle.dump(tfidf_matrix, f)

with open("tfidf_vectorizer.pkl", "wb") as f:
    pickle.dump(vectorizer, f)

print("TF-IDF representations saved successfully!")
```
