```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data_titanic.csv")
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.describe()
```

Plotting bar graph - provide x axis and yaxis

```python
plt.bar(data['Age'], data['Fare'])
plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Fare per Age")
plt.show()
```

```python
plt.bar(data["Survived"], data["Age"])
plt.xlabel("Survived")
plt.ylabel("Age")
plt.title("Age vs Survival")
```


Box Plot - 

```python
data.groupby("Sex")["Age"].mean()
```

```python
sns.catplot(x="Sex", y="Age", data=data, kind="box")
plt.show()
```

```python
sns.catplot(x="Parch", y ="Survived", data = data, kind="bar")

plt.show()
```

```python
sns.boxplot(x="Pclass",y="Age",data=data)
```

```python
sns.pairplot(data)
```

```python
sns.distplot(x = data['Age'], bins = 10)
```

```python
sns.violinplot(x=data['Sex'],y=data['Age'])
```

```python
sns.stripplot(x=data['Pclass'],y= data['Age'])
```
