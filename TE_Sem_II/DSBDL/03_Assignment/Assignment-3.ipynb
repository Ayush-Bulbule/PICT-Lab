**Operations on iris dataset**


```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris_df = pd.DataFrame(iris.data)

iris_df['class'] = iris.target

iris_df.columns = ['selpal_len','sepal_wid', 'petal_len','petal_wid', 'class']

iris_df.head()

```

```python
iris_df.info()
```

```python
iris_df['class'] = iris_df['class'].replace({0:'setosa',1:'versicolor',2:'virginica'})

iris_df.head()
```

```python
data_desc = pd.DataFrame(iris_df.groupby('class').describe())
data_desc
```

```python
data_desc.transpose()
```

```python
setosa_data = iris_df[iris_df['class'] =='setosa']
versicolor_data = iris_df[iris_df['class'] == 'versicolor']
viginica_data = iris_df[iris_df['class'] == 'verginica']
```
