```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('data_boston_housing.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

Remove all null values

```python
df = df.fillna(df.mean())
```

```python
plt.scatter(df['rm'], df['medv'], color="green")
plt.xlabel("Rooms per dewlling (RM)")
plt.ylabel('Median Value of Owner Occupied Homes')
plt.title('Scatter Plot : POSITION')
plt.show()
```

df.iloc[:, 0:-1]: Selects all rows (:) and all columns from the first (index 0) up to, but not including, the last column (index -1), effectively selecting all columns except the last one.
df.iloc[:, -1]: Selects all rows (:) and only the last column (index -1).

```python
# sns.heatmap(df.corr(), cmap="twilight", annot=True)
sns.heatmap(df.corr(), cmap="twilight")

```

```python
X = df.iloc[:,0:-1]
Y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.30,random_state=42) 
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train =  scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

<br/>
Training Data

```python
from sklearn.linear_model import LinearRegression
Lr = LinearRegression()

Lr.fit(X_train, Y_train)
```

```python
pred = Lr.predict(X_test)
pred
```

```python
from sklearn.model_selection import cross_val_score
MSE = cross_val_score(Lr, X_train,Y_train, scoring='neg_mean_squared_error', cv=10)
np.mean(MSE)
```

```python
from sklearn.metrics import r2_score 
score= r2_score(pred, Y_test)
score
```
